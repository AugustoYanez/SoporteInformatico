```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('datosencuesta.csv')



graf1 = df['Te preocupa que la inteligencia artificial pueda remplazar tu puesto de trabajo en el futuro?'].value_counts()


colors = ['#ff9999', '#66b3ff', '#99ff99']


plt.figure(figsize=(6, 6))
plt.pie(graf1, labels=graf1.index, autopct='%1.1f%%', colors=colors)
plt.title('Te preocupa que la IA pueda reemplazar tu puesto de trabajo en el futuro?')


plt.show()

# ---------------------------------------------------
graf2 = df['Cul es tu nivel de experiencia en el uso de tecnologas de inteligencia artificial?'].value_counts().sort_index()


plt.figure(figsize=(8, 6))
plt.plot(graf2.index, graf2.values, marker='o', color='skyblue', linestyle='-')
plt.xlabel('Nivel de Experiencia')
plt.ylabel('Nmero de Personas')
plt.title('Distribucin del Nivel de Experiencia en IA')


for x, y in zip(graf2.index, graf2.values):
    plt.text(x, y, str(y), ha='center', va='bottom')


plt.grid(True)
plt.show()

# ---------------------------------------------------

graf3 = df['Cmo tomas el impacto de la inteligencia artificial en la creatividad en tu rea de trabajo?'].value_counts()


plt.figure(figsize=(8, 6))
graf3.plot(kind='bar', color='skyblue')
plt.xlabel('Impacto en la Creatividad')
plt.ylabel('Nmero de Votantes')
plt.title('Nmero de Votantes por Impacto de la IA en la Creatividad')
plt.xticks(rotation=45)


for i, value in enumerate(graf3):
    plt.text(i, value + 0.5, str(value), ha='center')


plt.tight_layout()
plt.show()

# ---------------------------------------------------

graf4 = {
    'Nada familiarizado': 1,
    'Algo familiarizado': 2,
    'Bastante familiarizado': 3,
    'Muy familiarizado': 4
}
df['Familiaridad Numrica'] = df['Qu tan familiarizado ests con los conceptos bsicos de inteligencia artificial?'].map(graf4)

plt.figure(figsize=(8, 6))
plt.hist(df['Familiaridad Numrica'], bins=range(1, 6), color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Familiaridad')
plt.ylabel('Nmero de Personas')
plt.title('Familiaridad con Conceptos Bsicos de IA')
plt.xticks([1.5, 2.5, 3.5, 4.5], graf4.keys(), ha='center')
plt.grid(True)
plt.tight_layout()
plt.show()
```
